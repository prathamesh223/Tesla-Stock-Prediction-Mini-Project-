```python
#splitting the dataset into features(Independent Variables)and labels(Dependent variables).
X = df.drop(columns='next_day_close')
Y = df['next_day_close']
```

```python
print(X)
```

```python
print(Y)
```

## Spliting Dataset for Training and Testing

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape,)
```

```python
print(X_train,X_test)
```

```python
print(Y.shape,Y_train.shape,Y_test.shape,)
```

```python
print(Y_train,Y_test)
```

```python
#Standardization makes it easier to compare and interpret the coefficients or feature importance scores of different features within a model.
#Since all features are on the same scale, their coefficients or importance scores are directly comparable in terms of their relative 
#influence on the model predictions.
```

```python
from sklearn.preprocessing import StandardScaler
Scaler=StandardScaler()
X_train_scaled=Scaler.fit_transform(X_train)
X_test_scaled=Scaler.fit_transform(X_test)
```

```python
print(X_train_scaled.shape, X_test_scaled.shape)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train,Y_train)
lr_predict=lr.predict(X_test)
```

```python
#For regression tasks, mean absolute error (MAE) and mean squared error (MSE) are common evaluation metrics.
#You're already using mean_absolute_error, but you might also consider mean_squared_error for your regression evaluation.
```

```python
from sklearn.metrics import mean_absolute_error
lr_mean_abs_error=mean_absolute_error(Y_test,lr_predict)
lr_mean_abs_error
```

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dr=DecisionTreeRegressor()
```

```python
dr.fit(X_train,Y_train)
dr_predict=dr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
dr_mean_abs_error=mean_absolute_error(Y_test,dr_predict)
dr_mean_abs_error
```

## K-Nearest Neighbors (KNN) regressor

```python
from sklearn.neighbors import KNeighborsRegressor
kr=KNeighborsRegressor()
```

```python
kr.fit(X_train,Y_train)
kr_predict=kr.predict(X_train)
```

```python
from sklearn.metrics import mean_absolute_error
kr_mean_abs_error=mean_absolute_error(Y_test,dr_predict)
kr_mean_abs_error
```

## Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
```

```python
rfr.fit(X_train, Y_train)
rfr_prediction = rfr.predict(X_test)
```

```python
rfr_mean_abs_error = mean_absolute_error(Y_test, rfr_prediction)
rfr_mean_abs_error 
```

## Comparing The Mean Absolute Error Of Models

```python
Classifiers=['LinearRegression','Decision Tree Regressor','K-Nearest Neighbors regressor','Random Forest Regressor']
mean_abs_errors=[lr_mean_abs_error,dr_mean_abs_error,kr_mean_abs_error,rfr_mean_abs_error]
import matplotlib.pyplot as plt
plt.bar(Classifiers,mean_abs_errors,color=['cyan','red','green','magenta'])
plt.xticks(rotation=45)
plt.xlabel('Classifiers')
plt.ylabel('Mean_Absolute_Error')
plt.title('Comparing Mean Absolute Error Of Different Models ')
plt.show()
```
